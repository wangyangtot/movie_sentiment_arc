```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
path_to_load='/Users/yangyangwang/Documents/gutenberg/imsdb_scripts/normed_sentiment/'
```

```python
file_names=list(set(os.listdir(path_to_load)))
```

```python
sentiment_matrix=[]
index_name_dict={}
index=0
for file_name in file_names:
    sentiment=pd.read_csv(path_to_load+file_name).values
    sentiment=sentiment.ravel()
    sentiment_matrix.append(sentiment)
    index_name_dict[index]=file_name
    index+=1
    
```

```python
import numpy as np
from sklearn.cluster import KMeans
x=np.array(sentiment_matrix)
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(x)
```

```python
def ClusterIndices_group(clustNum, labels_array): #list comprehension
    return np.array([i for i, x in enumerate(labels_array) if x == clustNum])
```

```python
five=ClusterIndices_group(5, kmeans.labels_)
```

```python
five_sentiment = [sentiment_matrix[i] for i in five]
```

```python
for one in five_sentiment:
    
    one=np.asarray(one).ravel()
    plt.plot(one)
```
