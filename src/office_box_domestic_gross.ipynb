```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import os
import re
```

```python
def scraping_one_page(domestic_gross_path):
    movie_title=[]
    domestic_gross=[]
    response = requests.get(domestic_gross_path)
    soup = BeautifulSoup(response.text, 'html.parser')
    soup=soup.findAll('tr',bgcolor=re.compile('#f'))
    for movie_entry in soup:
        movie_description=movie_entry.findAll('b')
        movie_title.append(str(movie_description[0])[3:-4].replace(' ','-').lower())
        domestic_gross.append(str(movie_description[1])[3:-4])
    domestic_gross_one_page=pd.DataFrame({'movie_title':movie_title,'domestic_gross':domestic_gross},columns=[['movie_title','domestic_gross']])   
    return domestic_gross_one_page
```

```python
def scraping_all():
    domestic_gross_path_pre='https://www.boxofficemojo.com/alltime/domestic.htm?page='
    for page_num in range(1,159): ## total page is 158 by inspection
        print page_num
        domestic_gross_one_page=scraping_one_page(domestic_gross_path_pre+str(page_num)+'&p=.htm')
        if page_num==1: 
            domestic_gross_all=domestic_gross_one_page
        else: 
            domestic_gross_all=pd.concat([domestic_gross_all,domestic_gross_one_page])
    return domestic_gross_all
```

```python
if __name__ == '__main__':
    domestic_gross_all=scraping_all()
    domestic_gross_all.to_csv('../data/box_office/demostic_gross.csv',index=False)
```
