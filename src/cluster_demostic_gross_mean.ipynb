```python
import pandas as pd
import numpy as np
import os
import re
```

```python
cluster_name_file_path='../data/k_means_result/movie_name_cluster/'
total_movie_revenue_path='../data/select_total_movie_demostic_gross.csv'
movie_gross_cluster_path='../data/k_means_result/movie_gross_cluster/'
total_movie_revenue=pd.read_csv(total_movie_revenue_path)
```

```python
file_names=os.listdir(cluster_name_file_path)
```

```python
cluster_movie_revenue_list=[]
for file_name in file_names[1:]:
    print file_name[:-4]
    tem=np.load(cluster_name_file_path+file_name) 
    tem=[x[:-4] for x in tem.tolist()]
    if not os.path.exists(movie_gross_cluster_path):
        os.mkdir(movie_gross_cluster_path)
    cluster_movie_revenue=total_movie_revenue[total_movie_revenue['movie_title'].isin(tem)]
    #print cluster_movie_revenue['select_gross']
    cluster_movie_revenue_list.append(cluster_movie_revenue['select_gross'].mean())
    cluster_movie_revenue.to_csv(movie_gross_cluster_path+file_name[:-4]+".csv",index=False)
```
